```python
import numpy as np
import pandas as pd
import csv
```

```python
raw_df = pd.read_csv('Raw_eyes.csv',names = ['Wire ID','Meas_type','Gauge','Board_type','D3','D2','D1','D0','CMD0'])
```

```python
raw_df
```

```python
# Seperate the types:
lboard_df = raw_df[raw_df['Board_type']=='0'].copy()

lboard_df.head()
```

```python
yboard_df = raw_df[raw_df['Board_type']=='1'].copy()
yboard_df.head()
```

```python
#Clean Yellow Board Measurements
yboard_df.columns = ['Wire ID','Meas_type','Gauge','Board_type','D1','D0','CMD0','d1','d2']
yboard_df.drop(['d1','d2'],axis =1,inplace=True)
yboard_df.reset_index(drop=True, inplace=True)
yboard_df.head()
```

```python
lboard_df.drop(['D3','D2'],axis =1,inplace=True)
lboard_df.reset_index(drop=True, inplace=True)
lboard_df.head()
```

```python
trial_df = pd.concat([lboard_df,yboard_df])
trial_df.reset_index(drop=True, inplace=True)
trial_df
```

```python
trial_df.to_csv('CLean_eyes.csv', index=False)
```
